# **TRAIN-VAL SPLIT**

```python
RAW_DATA_DIR = Path("../data/train")
PROCESSED_DATA_DIR = Path("../data/processed")
TRAIN_RATIO = 0.85
VAL_RATIO = 0.15

CLASS_NAMES = {
    "0_Recyclable": "Recyclable",
    "1_Electronic": "Electronic",
    "2_Organic": "Organic",
}

print("Memulai split dataset")
split_and_organize_dataset(
    raw_dir=RAW_DATA_DIR,
    output_dir=PROCESSED_DATA_DIR,
    class_names=CLASS_NAMES,
    train_ratio=TRAIN_RATIO,
    seed=RANDOM_SEED,
)
```

```python
ENGLISH_CLASS_NAMES = sorted(CLASS_NAMES.values())

print(f"{'Kelas':12} | {'Train':>6} | {'Val':>6}")
print("-" * 32)

total_train, total_val = 0, 0
for cls in ENGLISH_CLASS_NAMES:
    n_train = len(list((PROCESSED_DATA_DIR / "train" / cls).glob("*")))
    n_val = len(list((PROCESSED_DATA_DIR / "val" / cls).glob("*")))
    total_train += n_train
    total_val += n_val
    print(f"{cls:12} | {n_train:6d} | {n_val:6d}")

print("-" * 32)
print(f"{'Total':12} | {total_train:6d} | {total_val:6d}")
```
